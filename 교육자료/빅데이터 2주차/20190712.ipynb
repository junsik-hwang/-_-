```python
headers = {"user-agent" : "Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"}
#headers = {"user-agent" : "bot"}
#param = {}
#400나오면 noun, 500나오면 다시 실행
import time

def download(method, url, param=None, data=None, timeout=1 , maxretires=3):
    try : 
        resp = requests.request(method, url, params=param, data=data, headers=headers)
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e :
        if 500 <= e.response.status_code < 600 and maxretires > 0:
            print(maxretires)
            time.sleep(timeout)
            resp = download(method, url, param, data, timeout, maxretires-1)
        else:
            print(e.response.status_code)
            print(e.response.reason)
    return resp
```

# Selector

- id 다음은 class 가 좋다.
- class는 . 으로 시작한다. (다중 상속 가능)

```python
html = """
    <html>
     <head></head>
      <body>
       <div id="result">
        <p class = "row">
         <a class = "red", id = "p_child1"> go to page1</a>
         <a class = "blue, id = "p_child2">go to page2</a>
         <div class = "blue"><a class = "red"> </a></div>
         <a class = "blue, id = "p_child2">go to page3</a>
        </p>
       </div>
      </body>
    </html>
"""
dom = BeautifulSoup(html, "lxml")
```

- 누구 밑에 있는 누구... 형식으로 실행함.

        dom.select()
        ID => # 어쩌고
        CLASS => .어쩌고
        태그 => 태그
        자손 => 공백
        자식 => >

```python
selector("div") => 2
selector("div#result") => 1
selector("#result") => 1
selector(".red") => 3
selector("body div") => 2 # div 2개 있어서 자손까지 긁어온다.
# #(태그) 까지 붙여서 완벽하게 지칭한다.
```
