```python
class B:
    def __repr__(self):
        return '졸리다'
    def __str__(self):
        return '진짜졸려'
```

```python
b = B()
```

```python
b
```

```python
print(b)
```

```python
class X:
    def __getattr__(self, attr):
        if attr == 'a':
            return '문근영2'
        return '문근영'
```

```python
a = X()
```

```python
a.a
```

```python
a.b
```
