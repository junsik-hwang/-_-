```python
# 타입 바꿔주는 명령어 : astype
data['수'] = data['수'].astype('int')
```

```python
data = data[data['수'] != 0]
```

```python
# 시 별로 연극/영화/극장이 몇 개 있나 찾아보기
data[data.iloc[:,-3] == '연극/영화/극장'].groupby('광역시도').count()
```

```python
# 서울시 구별로 부동산 업 이 몇개인가.
# size 는 미싱데이터 포함, count 는 미포함
data[(data['광역시도'] == '서울특별시') & (data['업종대분류'] == '부동산')].groupby(['시군구', '업종중분류']).count()
```

```python
# groupby 하면 split 하고 count 하는 순간 apply 가 지원된다. 
# < > 꺽새 보는 방법은? .count()
data[(data['광역시도'] == '서울특별시') & (data['업종대분류'] == '부동산')].groupby(['시군구', '업종중분류'])
```

## aggregation 3총사
- pd.crosstab()
- pivot_table()
- groupby

```python
# index 와 column 을 지정해주는 기능
pd.crosstab(data['광역시도'], data['년'], data['수'], aggfunc='count')
```

```python
pd.crosstab(data['년'], data['광역시도'], data['수'], aggfunc='count')
```

```python
data.groupby(['년', '광역시도']).count()
```

```python
# unstack() : column 과 index 를 왔다갔다.
data.groupby(['년', '광역시도']).count().unstack()
```

```python
# pivot_table : 입력을 강제한다. margins 로 부분합을 알려주는 것이 장점.
data.pivot_table('수','년', aggfunc='count', margins=True)
```
