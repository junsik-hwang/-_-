# apply , applymap, map

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips['ratio'] = tips.apply(lambda x:x['tip']/x['total_bill'], axis=1)
```

```python
tips
```

```python
# series 했을 때는 map 과 같지만
# 데이터 프레임에서는 이런 기능을 한다.
tips.apply(lambda x : x['total_bill']+x['tip'], axis=1)
```

```python
tips[['total_bill', 'tip']].apply(lambda x:sum(x))
```

```python
tips[['total_bill', 'tip']].applymap(lambda x:x+1)
```

```python
# 남자는 1, 여지는 2 로 바꾸기
# broadcasting 방법은 곤란하다. 그럴 때 쓰는 게 map
# map 안에 들어가는게 함수냐 , 딕셔너리냐...
tips.sex.map({"Male":1, "Female":2}) # 딕셔너리 버전
# tips.sex.map(lambda x : 1 if x == "Male" else 2) # 함수 버전
```

```python
tips[['total_bill', 'tip']].plot.bar()
```

```python
tips[['total_bill', 'tip']].plot(kind='bar')
```
